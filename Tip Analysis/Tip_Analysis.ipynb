```python
tip_day = df.groupby('day')['tip'].sum()
# Plot the pie chart
plt.figure(figsize=(8, 6))
colors = ['gold', 'skyblue', 'orange', 'green', 'plum']
plt.pie(tip_day, labels=tip_day.index, autopct='%1.1f%%', colors=colors)
plt.title('Tip Distribution by Day')
plt.legend(loc='best')
plt.show()
```

```python
# Group by day and sum the tips
day_tip = df.groupby('day')['tip'].sum()

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(day_tip.index, day_tip.values, color='teal')
plt.xlabel('Day of the Week')
plt.ylabel('Total Tips')
plt.title('Total Tips by Day')
#plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Group by smoker and time, then count the number of entries
smoker_time_counts = df.groupby(['smoker', 'time']).size()

# Combine labels like "Yes-Dinner", "No-Lunch", etc.
labels = [f"{smoker}-{time}" for smoker, time in smoker_time_counts.index]
sizes = smoker_time_counts.values

# Pie chart
plt.figure(figsize=(8, 6))
colors = ['gold', 'lightblue', 'tomato', 'lime']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Distribution by Smoker Status and Time')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Filter for Saturday and Sunday
weekend_data = df[df['day'].isin(['Sat', 'Sun'])]

# Group by day and sum tips
weekend_tip = weekend_data.groupby('day')['tip'].sum()

# Plot
plt.figure(figsize=(6, 5))
plt.bar(weekend_tip.index, weekend_tip.values, color=['orange', 'skyblue'])
plt.title('Total Tips on Saturday and Sunday')
plt.xlabel('Day')
plt.ylabel('Total Tips')
plt.tight_layout()
plt.show()
```

```python
# Group by day
tip_by_day = df.groupby('day')['tip'].sum()
count_by_day = df['day'].value_counts().sort_index()

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Total tips by day
axes[0].bar(tip_by_day.index, tip_by_day.values, color='coral')
axes[0].set_title('Total Tips by Day')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Tip Amount')

# Subplot 2: Number of entries per day
axes[1].bar(count_by_day.index, count_by_day.values, color='skyblue')
axes[1].set_title('Number of Entries by Day')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
